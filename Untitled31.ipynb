CODE FOR PLOTTING ANY GRAPH

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_graph(equation_str, x_range=(-10, 10), num_points=500):
    # Generate x values
    x = np.linspace(x_range[0], x_range[1], num_points)

    # Use eval to compute y values from the equation string
    try:
        y = eval(equation_str)
    except Exception as e:
        print(f"Error while evaluating the equation: {e}")
        return

    # Create the plot
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f"Graph of: {equation_str}")
    plt.title(f"Plot of the equation: {equation_str}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.axhline(0, color='black',linewidth=1)
    plt.axvline(0, color='black',linewidth=1)
    plt.legend()
    plt.show()

# Example usage
equation = "np.cos(2*x+6)"
plot_graph(equation)
```
